### Exercise 1: Create a Function to Calculate Farm Area

Write a function that:
- Takes length and width as parameters (in meters)
- Returns the area in hectares (1 hectare = 10,000 mÂ²)
- Test with: length=500m, width=400m

```python
# Your solution here
def farm_area (length, width):
    area_m2 = length * width
    area_hectares = area_m2 / 10000
    return area_hectares
length = 500
width = 400
result = farm_area (length, width)
print("="*50)
print(f"The farm area is {result} hectares.")
    
```

### Exercise 2: Create a Function with Multiple Returns

Write a function that:
- Takes total harvest (tons) and farm size (hectares)
- Returns: yield per hectare, performance rating
- Rating: "Excellent" if >5 tons/ha, "Good" if 3-5, "Poor" if <3
- Test with: harvest=120 tons, size=25 hectares

```python
# Your solution here
def farm_harvest (total_harvest, farm_size):
    yield_per_hectare = total_harvest / farm_size
    
    if yield_per_hectare > 5:
        print("Performance rating is Excellent")
    elif 3 <= yield_per_hectare <= 5:
        print("Performance rating is Good")
    else:
        print("Performance rating is Poor")

    return yield_per_hectare 

total_harvest = 120
farm_size = 25
yield_per_hectare = farm_harvest(total_harvest, farm_size) 
print(f"Total yield per hectare is {yield_per_hectare} hectares.")
```
